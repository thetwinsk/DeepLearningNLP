```python
from keras.preprocessing.text import Tokenizer
from numpy import array
from tensorflow.keras.utils import to_categorical
```

```python
def generate_seq(model, tokenizer, seed_text, n_words):
    in_text, result = seed_text, seed_text
    # generate a fixed number of words
    for _ in range(n_words):
        # encode the text as integer
        encoded = tokenizer.texts_to_sequences([in_text])[0]
        encoded = array(encoded)
        # predict a word in the vocabulary
        yhat = np.argmax(model.predict(encoded), axis=-1)
        # map predicted word index to word
        out_word = ''
        for word, index in tokenizer.word_index.items():
            if index == yhat:
                out_word = word
                break
        # append to input
        in_text, result = out_word, result + ' ' + out_word
    return result 
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts([data])
encoded = tokenizer.texts_to_sequences([data])[0]
```

```python
vocad_size = len(tokenizer.word_index) + 1
print('The size of vocabulary: %d' % vocad_size)
```

```python
sequences = list()
for i in range(1, len(encoded)):
    sequence = encoded[i-1:i+1]
    sequences.append(sequence)
print('The total of sequences: %d' % len(sequences))
```

```python
# print("Sequence: ", sequences)
```

```python
sequences = array(sequences)
X, y = sequences[:,0], sequences[:,1]
y = to_categorical(y, num_classes=vocad_size)
```

```python
# X, y
```

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM, Embedding
```

```python
model = Sequential()
model.add(Embedding(vocad_size, 10, input_length=1))
model.add(LSTM(50))
model.add(Dense(vocad_size, activation='softmax'))
print(model.summary())
```

```python
# compile network
model.compile(loss='categorical_crossentropy', optimizer='adam', 
              metrics=['accuracy'])
```

```python
#fit network
model.fit(X, y, epochs=500, verbose=2)
```

```python
# evaluate 
print(generate_seq(model, tokenizer, 'kiên', 8))
```

```python
# evaluate 
print(generate_seq(model, tokenizer, 'nay', 36))
```
